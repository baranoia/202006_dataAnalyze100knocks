```python
# ノック13 データに揺れがあるまま集計してみよう
```

```python
uriage_data["purchase_date"] = pd.to_datetime(uriage_data["purchase_date"])
uriage_data["purchase_month"] = uriage_data["purchase_date"].dt.strftime("%Y%m")
```

```python
uriage_data.head(3)
```

```python
res1 = pd.pivot_table(
    uriage_data , 
    index="purchase_month", 
    columns ="item_name",
    aggfunc="size",
    fill_value=0
)
res1
# 同じ商品なのに、表記の揺れのせいでべつべつにカウントされている(本来は26商品なのに99商品になっている)
```

```python
# pivotまだ慣れていなくてほんとにあっているか検算
res1.sum().sum(), len(uriage_data)
```

```python
res2 = pd.pivot_table(
    uriage_data , 
    index="purchase_month", 
    columns ="item_name",
    values="item_price",
    aggfunc="sum",
    fill_value=0
)
res2
```

```python
# pivotまだ慣れていなくてほんとにあっているか検算
res2.sum().sum(),  uriage_data["item_price"].sum()
```
